```python
from bs4 import BeautifulSoup
import pandas as pd  
from csv import writer
from csv import reader
import requests
import json

# reading the csv file
df = pd.read_csv("InitialwithSOCodeandGitCode.csv")

lists=[]
#print(len(lists))
#lists.append('SOCode')
#print(len(lists))
b=''
r=''
m=''

for x, y in df.iterrows():

    print(y.PostId)
    print(y.RevisionGUID)
    
    url='https://api.stackexchange.com/2.3/revisions/'+y.RevisionGUID+'?site=stackoverflow&filter=!1zIF476fd0uGMsa*ldEKh&key=NFl7EVxe0paITJEa0yzAaw(('
    s = requests.get(url)
    data = s.text
    parse_json = json.loads(data)
    #print(parse_json)
    #break
    for i,j in parse_json.items():
        print(i)
        if(i=='items'):
            #print("i inside item: ",i)
            #print("j: ",j[0]['body'])
            if(j[0]['post_id']==b and r==j[0]['revision_guid']):
                print("it's a match")
                lists.append(m)
            else:
                if('body' in j[0]):
                    print("newly formed ")
                    d=j[0]['body']
                    soup = BeautifulSoup(d)
                    s=[]
                    s=soup.findAll("code")
                    #print(s)
                    l = len(s)
                    m=''
                    #print(j.GHUrl)
                    for k in range(0,l,1):
                        m+=(s[k].renderContents().decode())+" "
                else:
                    print("empty")
                    m=''
                print(m)
                lists.append(m)

        b=y.PostId
        r=y.RevisionGUID
        print('b is now: ',b)
        print('r is now: ',r)
        break
    break




print(len(lists))

df['SOCode'] = lists
df.to_csv("./InitialwithSOCodeandGitCode.csv",index=False);
```
